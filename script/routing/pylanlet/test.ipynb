# Test LineString

```python
from linestring import LineString, LineType
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.array([[0, 0.9, 1, 2], [0, 0, -0.5, -1]]).T
l1 = LineString(1, points, LineType.WHITE_SOLID)
print(l1)
sample = l1.sample_points(0, 1, 100)
plt.scatter(points[:,0], points[:,1])
plt.plot(sample[:,0], sample[:,1])

project = np.array([[0.5, 0], [0.75, 0], [1.5, 0]])
projected = l1.project_point(project)
plt.scatter(project[:,0], project[:,1])
plt.scatter(projected[:,0], projected[:,1])
```
